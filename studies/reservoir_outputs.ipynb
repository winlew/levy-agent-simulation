**Objective**: Understand how the reservoir outputs look in dependence of the standard deviation and how it is generated through the neurons.

Note: This a showcase, do not alter the cells.

```python
import os
os.chdir(os.path.join(os.getcwd(), "../code"))
from tqdm import tqdm
from agent import Reservoir
import matplotlib.pyplot as plt
import numpy as np

TIME_STEPS = 1000
NUM_NEURONS = 1000
BURN_IN_TIME = 5000

def plot_reservoir_output(reservoir, standard_deviation, timesteps=TIME_STEPS):
    """
    Plot the output of the reservoir over time.
    """
    outputs = []
    for timestep in range(timesteps):
        outputs.append(reservoir.get_output(timestep))
    plt.figure(figsize=(25, 5))
    plt.plot(outputs, '.', label=f'Ïƒ={standard_deviation}')
    plt.ylim(-1.1, 1.1)
    plt.xlabel('Timestep')
    plt.ylabel('Reservoir Output')

def plot_neuron_activity(reservoir, n, timesteps=TIME_STEPS, num_neurons=NUM_NEURONS):
    """
    Plot neuron activity of first n neurons over time.
    """
    _, ax = plt.subplots(figsize=(25, 5))
    activity = reservoir.neuron_state_time_matrix
    for neuron in range(timesteps)[:n]:
        ax.scatter(range(timesteps), activity[:, neuron], label=f'Neuron {neuron}', alpha=0.5, s=7)
    ax.set_xlabel('Timestep')
    ax.set_ylabel('Neuron Activity')
    ax.legend(loc='upper right')
    ax.set_ylim(-1.1, 1.1)
    plt.tight_layout()
```

# Ordered 1000 Neuron Reservoir

```python
standard_deviation = 0.02
reservoir = Reservoir(TIME_STEPS, num_neurons=NUM_NEURONS, burn_in_time=BURN_IN_TIME, mean=0, standard_deviation=standard_deviation)
plot_reservoir_output(reservoir, standard_deviation)
plot_neuron_activity(reservoir, 3)
```

# Critical 1000 Neuron Reservoir

```python
standard_deviation = 0.032
reservoir = Reservoir(TIME_STEPS, num_neurons=NUM_NEURONS, burn_in_time=BURN_IN_TIME, mean=0, standard_deviation=standard_deviation)
plot_reservoir_output(reservoir, standard_deviation)
plot_neuron_activity(reservoir, 3)
```

```python
standard_deviation = 0.05
reservoir = Reservoir(TIME_STEPS, num_neurons=NUM_NEURONS, burn_in_time=BURN_IN_TIME, mean=0, standard_deviation=standard_deviation)
plot_reservoir_output(reservoir, standard_deviation)
plot_neuron_activity(reservoir, 3)
```
