**Objective**: Understand how the reservoir outputs look in dependence of the standard deviation.

```python
import os
os.chdir(os.path.join(os.getcwd(), "../code"))
from tqdm import tqdm
from agent import Reservoir
import matplotlib.pyplot as plt

TIME_STEPS = 200
NUM_NEURONS = 1000
BURN_IN_TIME = 100

standard_deviations = [0.01, 0.032, 0.08]

data = []

for std in tqdm(standard_deviations):
    outputs = []
    reservoir = Reservoir(TIME_STEPS, num_neurons=NUM_NEURONS, burn_in_time=BURN_IN_TIME, mean=0, standard_deviation=std)
    for timestep in range(TIME_STEPS):
        outputs.append(reservoir.get_output(timestep))
    data.append(outputs)

plt.figure(figsize=(16,8))

for i in range(len(standard_deviations)):
    plt.plot(data[i], '.', label=f'σ={standard_deviations[i]}')

plt.xlabel('Timestep')
plt.ylabel('Reservoir Output')
plt.title('Reservoir Outputs for Different Standard Deviations')
plt.legend()
```
